```python
df = pd.get_dummies(data,drop_first = True,dtype=int)
```

```python
independent = df[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent = df[["charges"]]
```

```python
independent.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=14)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
              'C':[10,100,1000,2000,3000],
              'gamma':['auto','scale']}

grid = GridSearchCV(SVR(),param_grid, refit=True,verbose = 3,n_jobs=-1)
grid.fit(X_train,y_train)
```

```python
re = grid.cv_results_
print(grid.best_params_)
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2score = r2_score(y_test,y_pred)
r2score
```
