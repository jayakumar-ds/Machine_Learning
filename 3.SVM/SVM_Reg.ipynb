```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("50_Startups.csv")
```

```python
data.head(4)
```

```python
df = pd.get_dummies(data,drop_first = True,dtype=int)
```

```python
df.head(5)
```

```python
df.columns
```

```python
independent = df[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent = df[["Profit"]]
```

```python
independent.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=14)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='sigmoid',C=0.2)
regressor = regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#Only for linear data weight works not for (rbf,poly,sigmoid)
#weight = regressor.coef_
#weight
```

```python
bias = regressor.intercept_
bias
```

```python
from sklearn.metrics import r2_score
r2score = r2_score(y_test,y_pred)
r2score
```
