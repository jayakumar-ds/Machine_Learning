```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("50_Startups.csv")
```

```python
data.head(5)
```

```python
data.columns
```

```python
# Using oneHot Encoding because it's a nominal data> converting categorical data to numerical data

df = pd.get_dummies(data,drop_first=True, dtype=int)
```

```python
df.head(5)
```

```python
df.columns
```

```python
independent = df[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent.head(5)
```

```python
dependent = df[["Profit"]]
```

```python
dependent.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Weight finding
weight = regressor.coef_
weight
```

```python
#Bias
bias = regressor.intercept_
bias
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2score = r2_score(y_test,y_pred)
```

```python
r2score
```

```python
# once model is finilized we need to dump/save the model for further use:
import pickle
filename = "finilized_mul_linear_reg.sav"
pickle.dump(regressor, open(filename,'wb'))
```

```python
#For sample testing reloaded the saved model to pridect new vaariable or features by passing new values

loaded_model = pickle.load(open("finilized_mul_linear_reg.sav",'rb'))
result = loaded_model.predict([[91234,108623.10,24930.22,1,0]])
```
