## **ADA Boost Regressor**

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read and store the csv file to "data" variable
data = pd.read_csv("insurance_pre.csv")
```

```python
data.head(5)
```

```python
df = pd.get_dummies(data,drop_first=True, dtype =int)
```

```python
df.head(5)
```

```python
independent = df[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = df[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=32)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score
```

```python
ada_reg = AdaBoostRegressor(learning_rate = 2.0,loss='linear',n_estimators=200,random_state=32)
ada_reg.fit(X_train,y_train)
```

```python
y_pred = ada_reg.predict(X_test)
```

```python
r2score = r2_score(y_test,y_pred)
r2score
```

## **XG Boost**

```python
from xgboost import XGBRegressor
```

```python
# Create XGBoost model
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)
```

```python
# Train the model
xgb.fit(X_train, y_train)

# Predictions
y_pred3 = xgb.predict(X_test)
```

```python
# Evaluate performance
r2 = r2_score(y_test, y_pred3)
print(f"R² Score: {r2:.4f}")
```

## **GradientBoosting Regressor**

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_reg = GradientBoostingRegressor(n_estimators=50,learning_rate=0.1,random_state=0)
gb_reg.fit(X_train,y_train)
```

```python
y_pred2 = gb_reg.predict(X_test)
```

```python
gb_reg_r2score = r2_score(y_test,y_pred2)
gb_reg_r2score
```
