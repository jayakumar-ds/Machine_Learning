## **RF-Grid-Without Training and Testing**

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("insurance_pre.csv")
data.head(5)
```

```python
df = pd.get_dummies(data,drop_first = True,dtype=int)
```

```python
independent = df[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent = df[["charges"]]
```

```python
independent.head(5)
```

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
'max_features': ['sqrt','log2'],
'n_estimators':[10,50,100,250,500],
'max_depth': [5, 10, None]}
```

```python
grid = GridSearchCV(RandomForestRegressor(),param_grid,refit=True,cv=5,verbose=3,n_jobs=-1)
grid.fit(dependent,independent)
```

```python
print(grid.best_params_)
```

```python
re=grid.cv_results_
table = pd.DataFrame.from_dict(re)
table
```

```python
table['rank_test_score'].min()
```

```python
table.loc[table['rank_test_score'] == table['rank_test_score'].min()]
```

```python
rf_Reg_Grid = RandomForestRegressor(criterion= 'squared_error', max_depth= 5, max_features= 'log2', n_estimators= 500)
rf_Reg_Grid.fit(independent,dependent)
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=rf_Reg_Grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
#change the paramter,play with it.
print("Future_Prediction={}".format(Future_Prediction))
```
