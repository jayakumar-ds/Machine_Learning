```python
#importing pandas for handling dataset and database
#matplotlib is used to plot the data's as we need or to find how the data is being distributed,
# also pie chart,violin plot, scatter plot, bar chart etc...,
import pandas as pd
import matplotlib.pyplot as plt
# warnings are used to handle/ignore some of the warning from the code, unwanted messages we can handle.
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the data from CSV File and saving it to data variable to reuse.
data = pd.read_csv("Salary_Data.csv")
```

```python
#head method to show first 5 rows of the data.
data.head(5)
```

```python
# Here I'm seperating the feature/variable and target variable with variable names.
independent = data[['YearsExperience']]
dependent = data[['Salary']]
```

```python
# From scikit family we are importing train_test_split function/method to split the data as training and testing data.
#Then assigning those splited data to 4 different variable name 
#Then using test_size - 70% data to training and 30% data to test
#random_state wil be like fixed data with some random no, so that next time when we are running the code it will execute the same sset of data without any change

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#From same sciket family we are importing linearmodel to train the data
#after importing then create object for LinearRegression class then save to regressor variable
#using that variable i'm calling fit method/function then inside im passing parameters of splited trained data

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Here, once the model is trained using that variable name we can get the slope/weight value and same as bais value

weight = regressor.coef_
```

```python
weight
```

```python
bias = regressor.intercept_
```

```python
bias
```

```python
# Here, i'm predicting the values for 30% test data and saving it to y_pred variable
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

### **Evaluating Metrics**

```python
#from metrics i'm importing r2_score method to find the percentage or score for the model prediction

from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

### **Model Saving using Pickle Function:**

```python
# using pickle class,we are saving the file to use future for deployment process
import pickle
filename = "finailized_model_linear.sav" 
pickle.dump(regressor,open(filename, 'wb')) # wb - write binary
```

### **Model Loading using Pickle Function:**

```python
# using same pickle class, i'm loading the saved file to predict the new values
loaded_model = pickle.load(open("finailized_model_linear.sav", 'rb')) #readbinary
result = loaded_model.predict([[13]]) # using loaded_model variable name we can predict new x values and im saving it to result variable
```
